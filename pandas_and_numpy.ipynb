```python
# creating numing areas

arr1 = np.array([1,2,3,4])
arr2 = np.array([[4 , 5, 6,] , [8 ,9 , 10]])

print("1D array:" , arr1)
print("2D array:" , arr2)

print("shape of 1D array:" , arr1.shape)
print("shape of 2D array:" , arr2.shape)

print("dimension of 1D array:" , arr1.ndim)
print("dimension of 2D array:" , arr2.ndim)

print("size of 1D array:" , arr1.size)
print("size of 2D array:" , arr2.size)

print(arr1.dtype)
print(arr2.dtype)

```
