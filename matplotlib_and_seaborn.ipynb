```python

import pandas as pd
import matplotlib.pyplot as plt

students = pd.read_csv('/content/StudentsPerformance.csv')
plt.plot(students['math score'].head(10))
plt.title('Math Scores Of First 10 Students')
plt.xlabel('Students Number')
plt.ylabel('Math Score')
plt.show()
```

```python
gender_counts = students['gender'].value_counts()
plt.bar(['Female','Male'] , gender_counts , color = ['Green' , 'Red'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(students['math score'],students['reading score'] , color = ['Black'])
plt.grid(True)
plt.xlabel('math')
plt.ylabel('reading')
plt.title('math vs reading')
plt.show()
```

```python
#histrogram
plt.hist(students['math score'] , bins = 10 , color = 'Black')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#pie chart
test_preparation_counts = students['test preparation course'].value_counts()
plt.pie(test_preparation_counts , labels = test_preparation_counts.index , autopct = '%1.1f%%')
plt.title('Test Preparation Course Distribution')
plt.show()
```

```python
#pie chart
test_preparation_counts = students['test preparation course'].value_counts()
plt.pie(test_preparation_counts , labels = ['none','completed'])
plt.title('Test Preparation Course Distribution')
plt.show()
```

```python
#multiple plot
#create 2 * 2 grid

fig,axes = plt.subplots(2,2 , figsize = (10,8))

#plot1 : histogram
axes[0,0].hist(students['math score'] , bins = 10 , color = 'Black')
axes[0,0].set_title('Math Score Distribution')

#bar chart
students['gender'].value_counts().plot(kind = 'bar', ax=axes[0,1])
axes[0,1].set_title('Gender Distribution')

#plot 3 : scatter
axes[0,1].scatter(students['math score'] [:50] ,students['reading score'] [:50], color = ['Black'])
axes[0,1].set_title('Math vs Reading Score')

#plot4 : line
axes[1,1,].plot(students['writing score'].head(10) , color = 'Black')
axes[1,1].set_title('writing scores')

plt.tight_layout()
plt.show()
```
